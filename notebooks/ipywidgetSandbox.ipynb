## Setting up a widget 

```python
#Had to pip install nodejs to get widgets to display in browser

from ipywidgets import widgets
slider = widgets.IntSlider(value=50)
slider
```

## Editing a widget's value 

```python
slider.value=100
```

```python
slider.value
```

## Listenting to Changes 

```python
from ipywidgets import HTML
text = HTML("The slider's value is <b>{}<b>".format(slider.value))
text
```

```python
def myfunc(change):
    text.value = "The slider's value is <b>{}<b>".format(slider.value)
    
slider.observe(myfunc, 'value')
```

## Plotting (see also ipyleaflet)

```python
import numpy as np

from sklearn import linear_model
from bqplot import LinearScale, Scatter, Figure
```

```python
# Styles
line_style = {'colors': ['magenta'], 'stroke_width':3}
scatter_style = {'colors': ['white'], 'stroke':'black'}
```

```python
x = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])
y = np.dot(x, np.array([1, 2])) + 3
print(x,y)
```

```python
regr = linear_model.LinearRegression()
regr.fit(x, y)
```

```python
pred = regr.predict(x)
print(pred)
```

```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np

def f(m, b):
    plt.figure(2)
    x = np.linspace(-10, 10, num=1000)
    plt.plot(x, m * x + b)
    plt.ylim(-5, 5)
    plt.show()

interactive_plot = interactive(f, m=(-2.0, 2.0), b=(-3, 3, 0.5))
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot
```

## Leaflet

```python
from ipywidgets import HTML
from ipyleaflet import Map, DrawControl, Marker, Popup
```

```python
!jupyter nbextension enable --py --sys-prefix ipyleaflet
```

```python
myMap = ipyleaflet.Map(center=[40.5, -107.5], zoom = 10)
dc = DrawControl(circle={'shapeOptions':{'color':'#0000FF'}})
myMap.add_control(dc)
myMap
```

```python
marker = Marker(location=[40.5, -107.5])
```

```python
myMap.add_layer(marker)
```

```python
message = HTML()
message.value = "Hello <b>World</b>"
message.placeholder = "Placeholder"
message.description = "Description"

# Popup associated to a layer
marker.popup = message
```

```python
marker.location
```

```python
def click_handler(marker):
    message.value = "Hello World 2"
marker.on_click(click_handler)
```
