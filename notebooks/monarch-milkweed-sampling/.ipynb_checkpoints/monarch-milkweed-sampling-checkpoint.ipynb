Create a model of milkweed abundance in a large area and sample using a simulated 50-m transect to estimate the sampling frequency required to detect milkweed at low abundances and the confidence intervals around milkweed densities when milkweed density is low.

1. Create a 2-d array where milkweed is distributed within cells at random according to the specified density
2. Write a function to simulate sampling milkweed with a 50-m transect
3. Bootstrap the sampling to model variability in density estimates at the specified density
4. Repeat for a range of densities and store confidence intervals of each

```python
import numpy as np
import pandas as pd
import math
import random
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
sns.set_style('white')
```

```python
class AssessmentArea:
    """A model of milkweed in an assessment area"""
    
    
    def __init__(self, identifier, area, density):
        self.identifier = identifier
        self.density = density
        self.area = area
        
    
    @property
    def milkweed_abundance(self):
        return self.area * self.density
    
    
    @property
    def dimensions(self):
        """Returns the length of the array in 1-m cells"""
        area_in_meters = self.area * 4048.856
        length_in_meters = math.sqrt(area_in_meters)
        length_in_cells = round(length_in_meters)
        return length_in_cells
    
    
    # Methods    
    def initialize(self):
        """Returns a 2-d array of 1-m cells where each cell with milkweed is represented by 1"""
        stems = np.asarray([1] * self.milkweed_abundance)
        pad_length = self.dimensions ** 2 - len(stems)
        one_stem_per_plot = np.pad(stems, (0, pad_length), 'constant', constant_values=0)
        np.random.shuffle(one_stem_per_plot)
        milkweed_array = one_stem_per_plot.reshape(self.dimensions, self.dimensions)
        return milkweed_array
    
    
    def sample(self, array, n=10, transect_length=50, plots_per_transect=10):
        """
        Simulates n 50-m transects collected within the Assessment Area with 10 plots
        per transect. Provide result of initialize method. Returns a list of values from 
        sampled plots
        """
        buffer = transect_length
        step = transect_length / (plots_per_transect - 1)
        aa_data = []
        for i in range(n):
            transect_data = []
            x = np.random.randint(0 + buffer, self.dimensions - buffer)
            y = np.random.randint(0 + buffer, self.dimensions - buffer)
            first_plot = array[x, y]
            transect_data.append(first_plot)
            direction = random.choice([(0,step), (step, 0), (0, -step), (-step, 0)])
            for i in range(plots_per_transect - 1):
                x = int(x + direction[0])
                y = int(y + direction[1])
                next_plot = array[x, y]
                transect_data.append(next_plot)
            aa_data.append(transect_data)
        aa_data = np.asarray(aa_data)
        return aa_data
    
    
    def sequential_sample(self, min_samples, max_samples, count_by):
        """Sequential sampling module. Returns dataframe."""
        self.min_samples = min_samples
        self.max_samples = max_samples
        self.count_by = count_by
        self.x_labels = range(min_samples, max_samples + count_by, count_by*2)
        
        trials = int((max_samples - min_samples) / count_by + 1)
        means = np.empty(trials)
        lower_cis = np.empty(trials)
        upper_cis = np.empty(trials)
        stdevs = np.empty(trials)
        misses = np.empty(trials)

        n_index = 0
        for n in range(min_samples, max_samples + count_by, count_by):
            repetitions = 100    
            sample_means = np.empty(repetitions)
            sample_lower_cis = np.empty(repetitions)
            sample_upper_cis = np.empty(repetitions)
            sample_misses = np.empty(repetitions)
            for i in range (repetitions):
                array = self.initialize()
                sample = self.sample(array, n)
                sample_means[i] = mw_per_acre(sample.sum(), n)
                if sample.sum() == 0:
                    sample_misses[i] = True
                else:
                    sample_misses[i] = False
            means[n_index] = sample_means.mean()
            lower_cis[n_index] = np.percentile(sample_means, 20)
            upper_cis[n_index] = np.percentile(sample_means, 80)
            stdevs[n_index] = np.std(sample_means)
            misses[n_index] = sample_misses.mean()
            n_index += 1

            outcomes = pd.DataFrame({'mean': means,
                             'upper_ci': upper_cis, 
                            'lower_ci': lower_cis,
                            'stdev': stdevs,
                            'misses': misses}, 
                            index = range(min_samples, max_samples + count_by, count_by))
            outcomes['mw_score_low'] = outcomes['lower_ci'].map(mw_score)
            outcomes['mw_score_high'] = outcomes['upper_ci'].map(mw_score)
        return outcomes
    
    
    def chart_results(self, milkweed_array, outcomes):
        fig, ax = plt.subplots(2, 2, figsize=(13,8))
        ax1 = ax[0,0]
        ax2 = ax[0,1]
        ax3 = ax[1,0]
        ax4 = ax[1,1]

        #Plot 1: Heatmap of milkweed presence
        sns.heatmap(milkweed_array, ax=ax1, cbar=False, cmap='Greens')
        ax1.set_axis_off()
        ax1.set_title(f'milkweed distribution ({self.density} stems per acre)')

        #Plot 2: Confidence intervals
        sns.barplot(data = outcomes, x = outcomes.index, y = 'upper_ci', color = "red", ax=ax2)
        sns.barplot(data = outcomes, x = outcomes.index, y = 'mean', color = "blue", ax=ax2)
        sns.barplot(data = outcomes, x = outcomes.index, y = 'lower_ci', color = "white", ax=ax2)
        ax2.set_ylabel('mw_density confidence intervals')
        ax2.axhline(y=self.density, color = 'white')
        
        ax2b = ax2.twinx()  # instantiate a second axes that shares the same x-axis
        ax2_ticks = ax2.get_yticks()
        ymin, ymax = ax2.get_ylim()
        ax2b.set_ylim((mw_score(ymin), mw_score(ymax)))
        ax2b.set_yticks(list(map(mw_score, ax2.get_yticks())))
        ax2b.set_ylabel('milkweed score (0 - 100)')
    

        #Plot 3: Standard Deviation as a function of n: number of transects
        ax3.plot(outcomes.index, outcomes['mean'], linestyle = '', color = 'blue', marker = 'x')
        ax3.set_xlabel('number of transects')
        ax3.set_ylabel('milkweed density (stems/acre)', color='blue')
        ax3.set_xticks(self.x_labels)
        ax3.tick_params(axis='y', labelcolor='blue')

        ax3.axhline(y=self.density, color = 'grey', linestyle = ':')

        ax3b = ax3.twinx()  # instantiate a second axes that shares the same x-axis

        color = 'red'
        ax3b.set_ylabel('stdev', color=color)
        ax3b.plot(outcomes.index, outcomes['stdev'], color = color, linestyle = '--')
        ax3b.tick_params(axis='y', labelcolor=color)

        #Plot 4: Proportion of false negatives (no milkweed when present) as a function
        # of n: number of transects
        ax4.plot(outcomes.index, outcomes['mean'], linestyle = '', color = 'blue', marker = 'x')
        ax4.set_xlabel('number of tranescts')
        ax4.set_ylabel('milkweed density (stems/acre)', color='blue')
        ax4.set_xticks(self.x_labels)
        ax4.tick_params(axis='y', labelcolor='blue')

        ax4.axhline(y=self.density, color = 'grey', linestyle = ':')

        ax4b = ax4.twinx()  # instantiate a second axes that shares the same x-axis

        color = 'red'
        ax4b.set_ylabel('proportion of Type II errors', color=color)
        ax4b.plot(outcomes.index, outcomes['misses'], color = color, linestyle = '--')
        ax4b.tick_params(axis='y', labelcolor=color)
        ax4b.set(ylim=(0, 1))

        fig.tight_layout()
        plt.rcParams.update({'font.size': 13})
        return plt 

def mw_per_acre(count, n, plots_per_transect=10):
    m2_per_acre = 4046.856
    sample_area = n * plots_per_transect
    return count / sample_area * m2_per_acre

def mw_score(x):
    if x >= 2000:
        return 1
    else:
        return round((1 - (2/ (1 + np.exp(6 * x/2000))))*100,2)
        
```

```python
aa_id = 1
aa_area = 50
aa_density = 1500
aa = AssessmentArea(aa_id, aa_area, aa_density)
milkweed_array = aa.initialize()
```

```python
min_transects = 5
max_transects = 100
count_by = 5
outcomes = aa.sequential_sample(min_transects, max_transects, count_by)
plot = aa.chart_results(milkweed_array, outcomes)
plot.show()
```
