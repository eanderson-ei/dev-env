```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fn = 'C:/Users/Erik/Downloads/pcm-events.csv'
df = pd.read_csv(fn)
df['Start Date of Training or Event'] = pd.to_datetime(df['Start Date of Training or Event'])
df.head()
```

```python
df.columns
```

```python
dimensions = [
    'PPL Workstream',
    'Type of Training or Event',
    'Training/ Event Delivery Format',
    'Client Office',
    'Country Where Training or Event Took Place'
]

measures = ['# \nfemale participants\nTOTAL', '# \nmale participants\nTOTAL']

no_dims = len(dimensions)

fig, axes = plt.subplots(no_dims, 1, figsize=(12,no_dims*5), sharex=True)

for dim, ax in zip(dimensions, np.ravel(axes)):
    df.groupby(dim)\
        .sum()[measures]\
        .plot(kind='barh', stacked=True, ax=ax)

plt.tight_layout()
plt.show()
```

```python
df.groupby([df['Start Date of Training or Event'].dt.strftime('%b'), df['Start Date of Training or Event'].dt.year])\
        .count()['PPL Workstream']\
        .plot(kind='bar', figsize=(12,8))
```
